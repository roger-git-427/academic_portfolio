```python
filepath = "data/01_raw/reservaciones_tcabdfront2.csv"
```

```python
import pandas as pd

df = pd.read_csv(filepath)

df.tail(10000)
```

```python
df.columns
```

```python
import pandas as pd

# Ruta donde Kedro buscará los datos de inferencia
inference_filepath = "data/01_raw/reservaciones_inference.csv"

df_filtrado = df[df["h_tot_hab"] > 0]
# Filtrar las últimas 100 filas
df_inference = df_filtrado.tail(100)

# Guardar el archivo CSV sin el índice
df_inference.to_csv(inference_filepath, index=False)
```

```python
df_inference.shape
```

```python
import numpy as np
import pandas as pd

# Reemplaza inf, -inf por NaN (por si los hay)
df_inference = df_inference.replace([np.inf, -np.inf], np.nan)

# Luego, reemplaza todos los NaN por un valor numérico válido (por ejemplo, 0 o -1)
df_inference = df_inference.fillna(0)  # puedes usar -1 si prefieres

# También puedes hacer esto si quieres solo limpiar columnas específicas
# df_inference["col1"] = df_inference["col1"].fillna(0)
df_inference.info()
```

```python
import requests

# Convertir a lista de diccionarios
data_dict = df_inference.to_dict(orient="records")
payload = { "data": data_dict }

# Enviar POST
url = "http://localhost:8080/predict"
response = requests.post(url, json=payload)

# Mostrar resultados
if response.status_code == 200:
    print("✅ Predicciones:")
    print(response.json())
else:
    print("❌ Error:", response.status_code)
    print(response.text)
```
